First, we have data of 6 wells, the depth with pressure values.

We are going to analyze and visualize the relationship between the depth and the pressure to estimate the O.W.C (Oil Water Contact), then we can represent the variability in OWC through distance between wells.

```python
# frist importing the most needed library.

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

Now we need to import the data (Excel sheet)

```python
# Load the data from Excel sheet into separate DataFrames

df_Well1 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well1')
df_Well2 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well2')
df_Well3 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well3')
df_Well4 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well4')
df_Well5 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well5')
df_Well6 = pd.read_excel('D:\D.T\geology\G412\The_4th_Project_in_Data_Analysis_With_Python.xlsx', sheet_name='Well6')
```

```python
# Extract relevant columns (depth and pressure) for the first four wells
Depth_1 = df_Well1['MD_1']
Pressure_1 = df_Well1['Pressure_1']

Depth_2 = df_Well2['MD_2']
Pressure_2 = df_Well2['Pressure_2']

Depth_4 = df_Well3['MD_4']
Pressure_4 = df_Well3['Pressure_4']

Depth_5 = df_Well4['MD_5']
Pressure_5 = df_Well4['Pressure_5']

# Extract relevant columns for the last two wells
Depth_1E = df_Well5['MD_1E']
Pressure_1E = df_Well5['Pressure_1E']

Depth_2E = df_Well6['MD_2E']
Pressure_2E = df_Well6['Pressure_2E']
```

```python
# Plot the data for the first four wells
plt.figure(figsize=(10, 6))
plt.plot(Pressure_1, Depth_1, 'o-', label='Well 1')
plt.plot(Pressure_2, Depth_2, 'o-', label='Well 2')
plt.plot(Pressure_4, Depth_4, 'o-', label='Well 3')
plt.plot(Pressure_5, Depth_5, 'o-', label='Well 4')
plt.xlabel('Pressure (PSI)')
plt.ylabel('Depth (M)')
plt.title('Pressure vs. Depth for Wells 1-4')
plt.legend()

plt.gca().invert_yaxis()  # Invert y-axis
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()

# Plot the data for the last two wells
plt.figure(figsize=(10, 6))
plt.plot(Pressure_1E, Depth_1E, 'o-', label='Well 5')
plt.plot(Pressure_2E, Depth_2E, 'o-', label='Well 6')
plt.xlabel('Pressure (PSI)')
plt.ylabel('Depth (M)')
plt.title('Pressure vs. Depth for Wells 5-6')
plt.legend()

plt.gca().invert_yaxis()  # Invert y-axis
plt.grid(True)  # Add gridlines
plt.legend()
plt.show()
```

Depending on the 1st graph, We will notice the sudden change in pressure (High Pressure) on the top which indicates the presence of a pressurized zone (The Cap-Rock), The 1st graph tells us the power of the water drive mechanism and it's strong aquifer support, While the 2nd graph represents Weak aquifer support.
